```python
import json
i=0
with open('purchase_log.txt', encoding='utf-8') as f:
    next(f)
    #испольуем next(f) чтобы не выводить первую строчку
    for line in f:
        line=line.strip()# испольщуем strip, чтобы удалитить пробелы вначале и в конце строки. 
        dict_=json.loads(line) # json для перевода списка со строками в словарь
        print(dict_['user_id'],dict_['category'])
        i+=1
        if i>1:
            break
```
